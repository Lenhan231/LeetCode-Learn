```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
df = pd.read_csv('datashet\\annual_gold.csv')
df.tail()
```

```python
# create a line chart
plt.figure(figsize=(15,7)) # 15 object x and 7 object y
sns.lineplot(
    data = df, 
    x = 'Date',
    y = 'Gold', 
    marker = 'D', 
    markersize = 10, 
    linestyle = '--') # visualize editing
plt.xticks(rotation = 30) # rotation for clearing the x
plt.show()
```
